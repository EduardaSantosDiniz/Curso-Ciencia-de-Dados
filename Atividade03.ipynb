```python
import pandas as pd
import numpy as np

df = pd.read_csv("movies.csv")
notas = pd.read_csv("ratings.csv")
df.columns = ["FilmeId2", "titulo", "Genero"]

df['Ano'] = df.titulo.str[-6:]
df['titulo'] = df.titulo.str[0:-6]
df['Ano']= df.Ano.str[1:-1]

notas.columns = ["UsuarioId", "FilmeId", "Nota", "Data"]
df['Notas'] = notas.Nota
df
```

```python
divisao = df[df['Ano'] == '2018']
divisao

media = divisao['Notas'].mean()
media
```
